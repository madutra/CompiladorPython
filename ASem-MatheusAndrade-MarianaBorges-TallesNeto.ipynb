```python
def simbolo(caractereAtual, estadoAnterior):
    if caractereAtual.isdigit():
        return 0
    elif caractereAtual == 'e' and estadoAnterior == 1 or caractereAtual == 'e' and estadoAnterior == 3:
        return 19  
    elif caractereAtual == 'E' and estadoAnterior == 1 or caractereAtual == 'E' and estadoAnterior == 3:
        return 20
    elif caractereAtual.isalpha():
        return 1
    elif caractereAtual == '"':
        return 2
    elif caractereAtual == '{':
        return 3
    elif caractereAtual == '}':
        return 4
    elif caractereAtual == ' ' or caractereAtual == '\n' or caractereAtual == '\t':
        return 5
    elif caractereAtual == 'EOF':
        return 6
    elif caractereAtual == '<':
        return 7
    elif caractereAtual == '>':
        return 8
    elif caractereAtual == '=':
        return 9
    elif caractereAtual == '-':
        return 10
    elif caractereAtual == '+':
        return 11
    elif caractereAtual == '*':
        return 12
    elif caractereAtual == '/':
        return 13
    elif caractereAtual == '.':
        return 14
    elif caractereAtual == '(':
        return 15
    elif caractereAtual == ')':
        return 16
    elif caractereAtual == ';':
        return 17
    elif caractereAtual == '_':
        return 21
    elif caractereAtual == ',':
        return 22
    else:
        return 18
```

```python
class Token:    #4.1 - Uma estrutura composta heterogênea (nó, registro, classe ...) denominada TOKEN.
    def __init__(self, Classe, Lexema, Tipo):
      self.__Classe = Classe.lower()
      self.__Lexema = Lexema
      self.__Tipo = Tipo

    def __repr__(self):
      rep = "Classe: {}, Lexema: {}, Tipo: {}".format(self.__Classe, self.__Lexema, self.__Tipo)
      return rep

    def get__Classe(self):
      return self.__Classe
    def set__Classe(self, classe):
      self.__Classe = str(classe).lower()

    def get__Lexema(self):
      return self.__Lexema
    def set__Lexema(self, lexema):
      self.__Lexema = str(lexema)

    def get__Tipo(self):
      return self.__Tipo
    def set__Tipo(self, tipo):
      self.__Tipo = str(tipo)
```

```python
def AnalisadorLexico():
    global cursor
    global linhaAtual
    global arquivo
    global numCaractereAtual
    global posicao
    global caractereAnterior
    global estadoAnterior
    global caractereAtual
    global tabelaDeSimbolos
    global production
    global error
    
    if cursor == numCaractereAtual:
        return Token("$","fim","fim")

    estadoAtual = 0
    estadoAnterior = 0
    bufferEntrada = ''
    caractereAtual = arquivo[cursor]
    flag = 0
    proximoEstado = tabelaTransicao[estadoAtual][simbolo(caractereAtual, estadoAnterior)]
    
    while proximoEstado != -1:
        estadoAtual = proximoEstado

        if caractereAtual == '\n':       
            linhaAtual += 1
            posicao = 1 
        
        if caractereAtual == '\t':       
            posicao += 1

        if caractereAtual != '\n' and caractereAtual != '\t': 
            bufferEntrada += caractereAtual
            if '{' in bufferEntrada:
                if '}' in bufferEntrada:
                    bufferEntrada = ''
                
        cursor += 1 
        posicao += 1 

        if cursor < numCaractereAtual: 
            caractereAtual = arquivo[cursor] 
            estadoAnterior = estadoAtual 
        else:
            flag = 1

        if flag != 1:
            proximoEstado = tabelaTransicao[estadoAtual][simbolo(caractereAtual, estadoAnterior)]
        else:
            proximoEstado = -1 
        
    if verificaEstadoFinal(estadoAtual):
        if bufferEntrada not in tabelaDeSimbolos and classe(estadoAtual) == 'id':
            tabelaDeSimbolos[bufferEntrada] = Token('id', bufferEntrada, None)
            tokenResult = Token(classe(estadoAtual),bufferEntrada,None)

        elif bufferEntrada not in tabelaDeSimbolos and classe(estadoAtual) == 'num':
            if estadoAtual == 3 or estadoAtual == 6: 
                tokenResult = Token(classe(estadoAtual),bufferEntrada,"real")
            else: 
                tokenResult = Token(classe(estadoAtual),bufferEntrada,"inteiro")
                
        elif bufferEntrada not in tabelaDeSimbolos and classe(estadoAtual) == 'lit':
            tokenResult = Token(classe(estadoAtual),bufferEntrada,"literal")
            
        elif bufferEntrada not in tabelaDeSimbolos and classe(estadoAtual) == 'ERRO':
            
            print('\t ERRO LÉXICO')
            print('\t O seguinte Caracter "' + bufferEntrada[len(bufferEntrada)-1] + '" não é um token reconhecido pelo analisador Léxico. ')
            print("\t Linha: "  + str(linhaAtual) + " | Posição: " + str(posicao-2))     
            error = True
            return AnalisadorLexico()       
        
        elif bufferEntrada not in tabelaDeSimbolos and classe(estadoAtual) == 'Comentário':
            return ''
        
        elif bufferEntrada not in tabelaDeSimbolos and classe(estadoAtual) == 'EOF':
            tokenResult = Token(classe(estadoAtual),bufferEntrada,None)
            
        elif bufferEntrada not in tabelaDeSimbolos and classe(estadoAtual) == 'OPR':
            tokenResult = Token(classe(estadoAtual),bufferEntrada,None)

        elif bufferEntrada not in tabelaDeSimbolos and classe(estadoAtual) == 'ATR':
            tokenResult = Token(classe(estadoAtual),bufferEntrada,None)
        
        elif bufferEntrada not in tabelaDeSimbolos and classe(estadoAtual) == 'OPA':
            tokenResult = Token(classe(estadoAtual),bufferEntrada,None)         
            
        elif bufferEntrada not in tabelaDeSimbolos and classe(estadoAtual) == 'AB_P':
            tokenResult = Token(classe(estadoAtual),bufferEntrada,None)
            
        elif bufferEntrada not in tabelaDeSimbolos and classe(estadoAtual) == 'FC_P':
            tokenResult = Token(classe(estadoAtual),bufferEntrada,None)
            
        elif bufferEntrada not in tabelaDeSimbolos and classe(estadoAtual) == 'pt_v':
            tokenResult = Token(classe(estadoAtual),bufferEntrada,None)
            
        elif bufferEntrada not in tabelaDeSimbolos and classe(estadoAtual) == 'VIR':
            tokenResult = Token(classe(estadoAtual),bufferEntrada,None)           
            
        elif bufferEntrada in tabelaDeSimbolos:
            tabelaSimbolosAux = palavrasReservadas()
            if bufferEntrada in tabelaSimbolosAux and bufferEntrada in tabelaDeSimbolos:
              tokenResult = Token(bufferEntrada,bufferEntrada,bufferEntrada)
            elif bufferEntrada not in tabelaSimbolosAux and bufferEntrada in tabelaDeSimbolos:
              tokenResult = Token(classe(estadoAtual),bufferEntrada,tabelaDeSimbolos[bufferEntrada].get__Tipo())           
      

    else:
        print('\t ERRO LÉXICO')
        print(f'\t {DefineErro(caractereAtual,estadoAtual)}')
        print("\t Linha: " + str(linhaAtual) + " | Posição: " + str(posicao-2) )
        bufferEntrada = ''
        error = True

        return AnalisadorLexico()
        
    caractereAnterior = bufferEntrada 
    return tokenResult
```

```python
def AnalisadorSemantico(t, tokenAtual, pilhaSemantico, cont_tab, contadorTab):
    global string_corpo
    global string_varTemp
    global contador_var_temp
    global error

    stringAux = ''

    # Verificações semânticas da produção gerada
    if t == 5:
      string_corpo += 3* '\n'
      
    elif t == 6: # D→ TIPO L pt_v
      pilhaSemantico[-2].set__Tipo(pilhaSemantico[-3].get__Tipo())
    
    elif t == 7: # L→ id vir L
      tokenAtual.set__Tipo(pilhaSemantico[-3].get__Tipo())
      pilhaSemantico[-1].set__Tipo(pilhaSemantico[-3].get__Tipo())
      Atualiza_tabela_simbolos(pilhaSemantico[-3])

    elif t == 8: # L→ id
      tipoinicial = pilhaSemantico[-1].get__Tipo()

      if pilhaSemantico[-4].get__Tipo() == 'inicio':
        pilhaSemantico[-1].set__Tipo('literal')
      else:
        pilhaSemantico[-1].set__Tipo(pilhaSemantico[-4].get__Tipo())

      if pilhaSemantico[-3].get__Tipo() == pilhaSemantico[-1].get__Tipo():
        AddObjeto(f"{pilhaSemantico[-3].get__Lexema()}; \n",1, cont_tab)
        if pilhaSemantico[-1].get__Tipo() == 'literal':
          varAux = 'literal'
        elif pilhaSemantico[-1].get__Tipo() == 'inteiro':
          varAux = 'int'
        elif pilhaSemantico[-1].get__Tipo() == 'real':
          varAux = 'double'
        AddObjeto(f"{varAux} {pilhaSemantico[-1].get__Lexema()}; \n",1, cont_tab)
      else:
        if pilhaSemantico[-1].get__Tipo() != pilhaSemantico[-3].get__Tipo and tipoinicial != 'literal':
          pilhaSemantico[-1].set__Tipo(pilhaSemantico[-3].get__Tipo())

        AddObjeto(f" {pilhaSemantico[-1].get__Lexema()}; \n",1, cont_tab)
      Atualiza_tabela_simbolos(pilhaSemantico[-1])
    
    elif t == 9: # TIPO→ inteiro
      tokenAtual.set__Tipo(pilhaSemantico[-1].get__Tipo())
      AddObjeto(f"int",1, cont_tab)

    elif t == 10: # TIPO→ real
      tokenAtual.set__Tipo(pilhaSemantico[-1].get__Tipo())
      AddObjeto(f"double",1, cont_tab)
    
    elif t == 11: # TIPO→ literal
      tokenAtual.set__Tipo(pilhaSemantico[-1].get__Tipo())
      AddObjeto(f"literal",1, cont_tab)
    
    elif t == 13:# ES→ leia id pt_v
      tokenAtual.set__Tipo(pilhaSemantico[-2].get__Tipo())
      if(pilhaSemantico[-2].get__Lexema() != '' and pilhaSemantico[-2].get__Classe() == 'id'):
        for token in tabelaDeSimbolos:
          if (tabelaDeSimbolos[token].get__Lexema() == pilhaSemantico[-2].get__Lexema()) and tabelaDeSimbolos[token].get__Tipo() is not None:
            pilhaSemantico[-2].set__Tipo(tabelaDeSimbolos[token].get__Tipo())
            Atualiza_tabela_simbolos(pilhaSemantico[-2])

      if(tabelaDeSimbolos[tokenAtual.get__Lexema()] != '' and pilhaSemantico[-2].get__Tipo() != None):
        stringAux = f'scanf('
        if(pilhaSemantico[-2].get__Tipo() == 'literal'):
          stringAux += '"%[^\\n]s", &' 
        elif(pilhaSemantico[-2].get__Tipo() == 'inteiro'):
          stringAux += '"%d", &'
        elif(pilhaSemantico[-2].get__Tipo() == 'real'):
          stringAux += '"%f", &'
        stringAux += f'{(pilhaSemantico[-2].get__Lexema()).replace(" ","")});\n'

        AddObjeto(stringAux, 1, cont_tab)      
      else:
        print(f'Erro leia: Variável não declarada.\n Linha:{linhaAtual} Coluna:{posicao-2} \n')
        error = True

    elif t == 14: #ES→ escreva ARG pt_v

      if(pilhaSemantico[-2].get__Tipo() is None):
        for token in tabelaDeSimbolos:
          if (tabelaDeSimbolos[token].get__Lexema() == pilhaSemantico[-2].get__Lexema() or (tabelaDeSimbolos[token].get__Lexema()).replace(" ","") == pilhaSemantico[-2].get__Lexema() or (f' {tabelaDeSimbolos[token].get__Lexema()}') == pilhaSemantico[-2].get__Lexema()) and tabelaDeSimbolos[token].get__Tipo() is not None:
            pilhaSemantico[-2].set__Tipo(tabelaDeSimbolos[token].get__Tipo())
            Atualiza_tabela_simbolos(pilhaSemantico[-2])

      if(tabelaDeSimbolos[tokenAtual.get__Lexema()] != '' and pilhaSemantico[-2].get__Tipo() != None):
        if( 'literal' in pilhaSemantico[-2].get__Tipo() ):
          stringAux = '"%s", ' 
        elif(pilhaSemantico[-2].get__Tipo() == 'inteiro'):
          stringAux = '"%d", '
        elif(pilhaSemantico[-2].get__Tipo() == 'real'):
          stringAux = '"%f", '
        
        if stringAux != '':
          stringAux += f'printf({stringAux}{pilhaSemantico[-2].get__Lexema()});\n'
          AddObjeto(stringAux, 1, cont_tab)      

    elif t == 15:
      tokenAtual = pilhaSemantico[-1]
      

    elif t == 16:
      tokenAtual = pilhaSemantico[-1]
      

    elif t == 17:
      if pilhaSemantico[-2].get__Tipo() is not None and pilhaSemantico[-2].get__Tipo() != '':
        tokenAtual = pilhaSemantico[-1]
        Atualiza_tabela_simbolos(pilhaSemantico[-1])      
      else:
        print(f'Erro: Variável não declarada.\n Linha:{linhaAtual} Coluna:{posicao-2} \n')
        error = True        

    elif t == 19:
      if(tokenAtual.get__Lexema() != '' and tokenAtual.get__Classe() == 'id'): #verifica se token é id ???? pra que meu pai
        for token in tabelaDeSimbolos:
          if (tabelaDeSimbolos[token].get__Lexema() == tokenAtual.get__Lexema() or (tabelaDeSimbolos[token].get__Lexema()).replace(" ","") == tokenAtual.get__Lexema()) and tabelaDeSimbolos[token].get__Tipo() is not None:
            tokenAtual.set__Tipo(tabelaDeSimbolos[token].get__Tipo())
            Atualiza_tabela_simbolos(tokenAtual)

      if tokenAtual.get__Tipo() is not None and pilhaSemantico[-4].get__Tipo() != '': # Verifica se id foi declaro, ata agora sim, id é [-4] 
        if pilhaSemantico[-2].get__Tipo() == pilhaSemantico[-4].get__Tipo(): #verifica de id.tipo = L.tipo ok
          AddObjeto(f'{pilhaSemantico[-4].get__Lexema()} = {pilhaSemantico[-2].get__Lexema()}; \n',1, cont_tab)
        else:
          error = True
          print(f'Erro: Tipos diferentes para atribuição.\n Linha:{linhaAtual} Coluna:{posicao-2} \n')
      else:
        error = True
        print(f'Erro: Variável não declarada.\n Linha:{linhaAtual} Coluna:{posicao-2} \n')

    elif t == 20:
      if pilhaSemantico[-1].get__Tipo() == pilhaSemantico[-3].get__Tipo():
        contador_var_temp += 1
        tabelaDeSimbolos[f'T{contador_var_temp}'] = Token('id', f'T{contador_var_temp}', 'inteiro')
        AddObjeto(f'int T{contador_var_temp};',2, cont_tab)
        AddObjeto(f'T{contador_var_temp} = {pilhaSemantico[-3].get__Lexema()} {pilhaSemantico[-2].get__Lexema()} {pilhaSemantico[-1].get__Lexema()}; \n',1, cont_tab)
        tokenAtual = tabelaDeSimbolos[f'T{contador_var_temp}']
      else:
        error = True
        print(f'Erro: Operandos com tipos incompatíveis.\n Linha:{linhaAtual} Coluna:{posicao-2} \n')

    elif t == 21:
      tokenAtual = pilhaSemantico[-1]

    elif t in (22,23):
      if(pilhaSemantico[-1].get__Lexema() != '' and pilhaSemantico[-1].get__Classe() == 'id'):
        for token in tabelaDeSimbolos:
          if (tabelaDeSimbolos[token].get__Lexema() == pilhaSemantico[-1].get__Lexema() or (tabelaDeSimbolos[token].get__Lexema()).replace(" ","") == pilhaSemantico[-1].get__Lexema()) and tabelaDeSimbolos[token].get__Tipo() is not None:
            pilhaSemantico[-1].set__Tipo(tabelaDeSimbolos[token].get__Tipo())
            Atualiza_tabela_simbolos(pilhaSemantico[-1])

      if pilhaSemantico[-1].get__Tipo() is not None:
        tokenAtual = pilhaSemantico[-1]
      else:
        if t == 22:
          error = True
          print(f'Erro: Variável não declarada.\n Linha:{linhaAtual} Coluna:{posicao-2} \n')

    elif t == 25:
      stringParaTab = contadorTab * '\t'
      AddObjeto(f'{stringParaTab}' + '} \n',1, cont_tab)
      contadorTab = contadorTab - 1
    elif t == 26:
      stringParaTab = contadorTab * '\t'
      AddObjeto(f'{stringParaTab}if({pilhaSemantico[-3].get__Lexema()}) \n',1, cont_tab)
      AddObjeto('{ \n \t',1, cont_tab)
      contadorTab = contadorTab + 1

    elif t == 27:
      if pilhaSemantico[-1].get__Tipo() == pilhaSemantico[-3].get__Tipo():
        contador_var_temp += 1
        tabelaDeSimbolos[f'T{contador_var_temp}'] = Token('id', f'T{contador_var_temp}', 'inteiro')
        AddObjeto(f'int T{contador_var_temp};',2, cont_tab)
        AddObjeto(f'T{contador_var_temp} = {pilhaSemantico[-3].get__Lexema()} {pilhaSemantico[-2].get__Lexema()} {pilhaSemantico[-1].get__Lexema()}; \n',1, cont_tab)
        tokenAtual = tabelaDeSimbolos[f'T{contador_var_temp}']
      else:
        error = True
        print(f'Erro: Operandos com tipos incompatíveis.\n Linha:{linhaAtual} Coluna:{posicao-2} \n')

    else:
      pass
    
    return tokenAtual
```

```python
def Atualiza_tabela_simbolos(id):
    if(id.get__Lexema() in tabelaDeSimbolos):
        tabelaDeSimbolos[id.get__Lexema()] = Token(id.get__Classe(), id.get__Lexema(), id.get__Tipo())

    if(f' {id.get__Lexema()}' in  tabelaDeSimbolos):
        tabelaDeSimbolos[f' {id.get__Lexema()}'] = Token(id.get__Classe(), id.get__Lexema(), id.get__Tipo())

    else:
      pass
```

```python
def AddObjeto(string, tipo, cont_tab):
    global string_corpo
    global string_varTemp
    if("}" in string):
      cont_tab -= 1
    
    if(tipo == 2):
      string_varTemp += "\t" + string + "\n"
    else:
      string_corpo += cont_tab * "\t" + string

    if("{" in string):
      cont_tab += 1
```

```python
def IniciaObjeto():

    arquivo = open('meuarquivo.c', 'w')

    objetoFinal = '#include<stdio.h> \n'
    objetoFinal += 'typedef char literal[256]; \n'
    objetoFinal += 'void main(void) \n'
    objetoFinal += '{   '
    objetoFinal += '/*----Variaveis temporarias----*/ \n'
    objetoFinal += string_varTemp
    objetoFinal += '/*------------------------------*/\n'
    objetoFinal += string_corpo + '\n'
    objetoFinal += '}'

    arquivo.write(objetoFinal)

    arquivo.close()

    return objetoFinal
```
